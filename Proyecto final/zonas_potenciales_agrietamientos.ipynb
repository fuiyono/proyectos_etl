```python
import geopandas as gdp
```

```python
gdf = gdp.read_file('data/inestabilidad_laderas/zonas_potenciales_de_agrietamiento.geojson')

print(gdf.head())

gdf.plot()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))

gdf.plot(
    column='magni_num',
    ax=ax,
    cmap='viridis',
    linewidth=1,
    edgecolor='w',
    legend=True,
)

ax.axis('off')

plt.show()
```

```python
import folium 

width = 800
height = 600

m = folium.Map(location=[19.4210812, -99.0413559], zoom_start=10, width=width, height=height)

folium.GeoJson(
    gdf,
    name='Zonas potenciales de agrietamientos'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
