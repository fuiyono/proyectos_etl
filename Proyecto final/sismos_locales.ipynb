## Peligro por sismos locales de la delagación Alvaro Obregon

En este notebook se analizará el peligro por sismos locales de la delegación Alvaro Obregon. Los datos se encuentran en un archivo GeoJSON que contiene información sobre el peligro por sismos locales en la delegación Alvaro Obregon.

```python
import geopandas as gpd
```

```python
# Cargar el archivo GeoJSON
gdf = gpd.read_file('data/alvaro_obregon/riesgo_por_sismos_locales.geojson')

# Ver las primeras filas
print(gdf.head())

# Ver información general
print(gdf.info())

# Ver el sistema de coordenadas (CRS)
print(gdf.crs)
```

## Exploración de los datos

```python
# Inspeccion
print(gdf.geom_type.value_counts())
```

## Resumen Estadístico de los Datos

```python
print(gdf.describe())
```

```python
print(gdf.isnull().sum())
```

## Visualización Incial

```python
# Graficar la geometría con Matplotlib
import matplotlib.pyplot as plt

gdf.plot(figsize=(10, 6), cmap="viridis", edgecolor="black")
plt.title("Mapa de Datos GeoJSON")
plt.show()
```

### Ver distribución de una columna numérica

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(gdf["longitud"], bins=20, kde=True)  # Reemplaza "INTENSIDAD" por tu columna numérica
plt.title("Distribución de LONGITUD")
plt.show()
```
